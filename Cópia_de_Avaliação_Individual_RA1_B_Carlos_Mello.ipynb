# Questão 1: 
O valor de um computador deprecia exponencialmente com o tempo. Seu valor inicial é de R$ 4.000, e a taxa de depreciação anual é de 20%. Qual será o valor do computador após 5 anos? Use a fórmula:

$$
V(t) = V_0 \cdot (1 - r)^t
$$

onde $V_0$ é o valor inicial, $r$ é a taxa de depreciação e $t$ é o tempo em anos. Não se esqueça de traçar um gráfico mostrando o o valor do bem depreciado, ano a ano, nos próximos 15 anos.

$$
V(5) = 4000 \cdot (1 - 0.20)^5 = 1310.72
$$

Em 5 anos o notebook terá o valor de R$ 1310.72

```python
expr = 4.000 * (1.0 - 0.20)**5
display(expr)
```

**Gráfico** 

```python
# Mostra os anos de 0 a 15
t = range(15)

# Valor inicial
V0 = 4000

# Taxa de depreciação anual
r = 0.2 

# Calcula o valor do computador ao longo dos 15 anos usando List Comprehension
V = [V0 * (1 - r)**t_i for t_i in t]

# Plota o gráfico
plt.plot(t, V, 'o--')
plt.xlabel('Tempo (anos)')
plt.ylabel('Valor (R$)')

plt.title('Depreciação do notebook')
plt.grid()
plt.show()

# Mostrando os valores de cada ano
for i in range(len(V)):
  print("Ano:" , i, "Valor:", V[i])
```

**Gráficos com valores**

```python
# Mostra os anos de 0 a 15
t = range(15)

# Valor inicial
V0 = 4000

# Taxa de depreciação anual
r = 0.2 

# Calcula o valor do computador ao longo dos 15 anos usando List Comprehension
V = [V0 * (1 - r)**t_i for t_i in t]

# Plota o gráfico
plt.plot(t, V, 'o--')
plt.xlabel('Tempo (anos)')
plt.ylabel('Valor (R$)')
plt.title('Depreciação do notebook')
plt.grid()

# Mostrando os valores
for t, V in zip(t, V):
  plt.text(t-0.2, V-0.1, '(Ano:{})'.format(t))

plt.show()
```

# Questão 2
Usando, a seu critério, as bibliotecas do Python, crie um gráfico para cada uma das funções a seguir. No label do gráfico deverá estar a função impressa com qualidade matemática (latex) e uma label dizendo se a função é crescente ou decrescente. 

1. $3^{x-1} - 2^x = 7$

2. $\log_5(2x + 1) + \log_5(x - 2) = 2$

**1**

```python
# Os valores para o gráfico
x = np.linspace(-10, 10, 100)
y = (3**(x-1) - 2**x)

# Plotagem do gráfico
fig, ax = plt.subplots()
ax.plot([-10,10], [2, 2])
ax.plot(x, y)

#Configura os eixos 
ax.axhline(y = 0, color = 'black')
ax.axvline(x = 0, color = 'black')

# Limite dos eixos
ax.set_xlim([-7, 7])
ax.set_ylim([-15, 15])

# Label e Título com Latex
plt.title('$3^{x-1} - 2^x = 7$. Função crescente')
plt.grid()

plt.show()
```

**2**

```python
# Os valores para o gráfico
x = np.linspace(-10, 10, 100)
y = np.log(5 * ((2*x) + 1)) + np.log(5 * (x - 2)) - 2

# Plotagem do gráfico
fig, ax = plt.subplots()
ax.plot([-10,10], [2, 2])
ax.plot(x, y)

#Configura os eixos 
ax.axhline(y = 0, color = 'black')
ax.axvline(x = 0, color = 'black')

# Label e Título com Latex
plt.title('$\log_5(2x + 1) + \log_5(x - 2) = 2$. Decrescente')
plt.grid()

plt.show()
```

1. $\lim_{x \to 0} \frac{\sin(2x)}{x}$

2. $\lim_{x \to \infty} \frac{e^x}{x^3}$

**1**

$\lim_{x \to 0} \frac{\sin(2x)}{x} = \lim_{x \to 0}\sin(2x) = 0  = \lim_{x \to 0}x = 0$
Precisamos fatorar a expressão.

$\lim_{x \to 0} \frac{\sin(2x)}{x} = \lim_{x \to 0} \frac{\sin(2x)}{x} \times \frac{2}{2} = 2 \times \lim_{x \to 0}\frac{\sin(2x)}{2x} = 2 \times \lim_{t \to 0}\frac{\sin(t)}{t} = 2 \times 1 = 2$

```python
# Define as váriaveis
x = np.linspace(-4.7, 4.7, 100)
y = np.sin(2*x) / x

# Plota o gráfico
plt.plot(x, y)
# Plota o valor do limite
plt.scatter(0,2, color = 'red')

plt.xlabel('X')
plt.ylabel('Y')

#Configura os eixos 
plt.axhline(y = 0, color = 'black')
plt.axvline(x = 0, color = 'black')
plt.grid()

plt.show()
```
